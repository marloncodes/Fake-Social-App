```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import svm
from  sklearn.metrics import classification_report

%matplotlib inline
```

```python
X,y = load_iris(return_X_y=True)
```

```python
iris = load_iris()
iris_pd = pd.DataFrame(data=X, columns=iris.feature_names)
iris_pd['target_name'] = pd.Categorical.from_codes(y,iris.target_names)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
iris_pd
```

```python
iris_pd.plot.scatter('petal length (cm)','petal width (cm)',c=y ,cmap='cividis')
```

```python
sns.pairplot(iris_pd[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)','target_name']],hue='target_name')
```

```python
clf = svm.SVC(C=1.0)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_test
```

```python
#AcurÃ¡cia
clf.score(X_test,y_test)
```

```python
print(classification_report(y_test,y_pred,target_names=iris.target_names))
```
